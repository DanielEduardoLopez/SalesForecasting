### **1.1 Background** <a class="anchor" id="intro"></a>

In the ever-changing world of today, it is of the utmost importance to be able to quantify the impact of the external economic indicators in the revenue of an organization. 

In the context of Mexico, it is possible to collect public data from different government offices such as INEGI or BANXICO, or from international sources such as the S&P500. 

In this sense, the purpose of the present proyect is to predict several economic indicators over the next few years and, then, use them to predict later the sales of AstraZeneca.
